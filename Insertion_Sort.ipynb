# 2. Basic Approach 


Time complexity = O(n^2)

```python
#Sort the array using insertion sort

class Solution:

    #Function to sort the list using insertion sort algorithm.    
    def insertionSort(self, alist, n):
        #code here
        for i in range(1,n):
            key = alist[i]
            j=i-1
            while j>=0 and key < alist[j]:
                alist[j+1]=alist[j]
                j-=1
            alist[j+1]=key
                
                    
            

#{ 
#  Driver Code Starts
if __name__=="__main__":
    t=int(input())
    for i in range(t):
        n=int(input())
        arr=list(map(int,input().split()))
    
        Solution().insertionSort(arr,n)
    
        for i in range(n):
            print(arr[i],end=" ")
    
        print()
# } Driver Code Ends
```

```python
#Sort the array using insertion sort

class Solution:
    def binary_search(self,alist,key,low,high):
        
        while low <= high:
            mid = low + (high-low)//2
            if alist[mid]==key:
                return mid+1
            elif key>alist[mid]:
                low = mid+1
            else:
                high = mid-1
        return low

    #Function to sort the list using insertion sort algorithm.    
    def insertionSort(self, alist, n):
        #code here
        for i in range(1,n):
            j=i-1
            key = alist[i]
            loc = self.binary_search(alist,key,0,j)
            #print(loc,i,j)
            
            #below line of code cann't be used as alist created here is local not global as x=x[:] makes copy of list
            #alist = alist[:loc] + [key] + alist[loc:i] + alist[i+1:]
            k=i-1
            while k>=loc:
                #print(alist[k],alist[k+1])
                alist[k+1]=alist[k]
                #print(alist[k],alist[k+1])
                k-=1
            
            alist[loc]=key
            
            #print(alist)    
                    
            

#{ 
#  Driver Code Starts
if __name__=="__main__":
    t=int(input())
    for i in range(t):
        n=int(input())
        arr=list(map(int,input().split()))
    
        Solution().insertionSort(arr,n)
    
        for i in range(n):
            print(arr[i],end=" ")
    
        print()
# } Driver Code Ends
```
